```python
from neural_avalanche_utac import NeuralAvalancheUTAC, GAMMA_BRAIN
print(f'Theoretical Γ_brain = {GAMMA_BRAIN}')
```

```python
# Instantiate the model
model = NeuralAvalancheUTAC(n_neurons=300, seed=42, segment_s=60.0)

# Run a 10-minute (600s) UTAC cycle
result = model.run_cycle(duration_seconds=600.0)
result
```

```python
# CREP state — the four diagnostic components
crep = model.get_crep_state()
print(f'C (autocorrelation):   {crep["C"]:.3f}')
print(f'R (power-law proxy):   {crep["R"]:.3f}')
print(f'E (emergence):         {crep["E"]:.3f}')
print(f'P (permutation ent.):  {crep["P"]:.3f}')
print(f'Γ (CREP tensor):       {crep["Gamma"]:.4f}  (theory: 0.251)')
print(f'σ_b (branching ratio): {crep["sigma_b"]:.4f}')
```

```python
# UTAC state
utac = model.get_utac_state()
print('UTAC state:', utac)
print(f'Is critical? {model.is_critical()}')
```

```python
# Phase events (avalanche clusters)
events = model.get_phase_events()
print(f'Total phase events recorded: {len(events)}')
if events:
    print('First event:', events[0])
```

```python
# Cross-domain universality check
univ = model.gamma_universality_check()
print(univ['interpretation'])
```

```python
# Zenodo record (metadata for open science)
record = model.to_zenodo_record()
print('Title:', record['title'])
print('DOI:', record['doi'])
print('Universality confirmed:', record['gamma_universality_confirmed'])
```
